```r
#Read Data 
vehicle <- read.csv("vehicle.csv", header= TRUE)
# HEADER true since we have variable name in the first row 
```

```r
head(vehicle)
```

```r
#structure of the data 
str(vehicle)
```

```r
#summary 
summary(vehicle)
```

```r
#labour hours has a minumum value of zero and labour cost 
#also zero so this seeems like a missing data. 
# we have to handle missing observations 
```

```r
vehicle$lh[vehicle$lh==0] <- mean(vehicle$lh)
vehicle$lc[vehicle$lc==0] <- mean(vehicle$lc)
# vehicle$lh[vehicle$lh==0] <- NA
# vehicle$lc[vehicle$lc==0] <- NA
summary(vehicle)
```

```r
# make a plot, we are interested in only mileage , lh and lc 
pairs(vehicle[3:5])
```

```r
# Get the corellation between variable 
cor(vehicle[3:5])
# we cannot seem to get the correlation whenever there is a missing data set
# we therefore give the missing values a mean value 
```

```r
# We have a high correlation between labour cost and 
# labour hours 
```

```r
# Data Partition 
set.seed(1234) #specific starting point for randomization
# split data into 70% and 30%
ind <- sample(2, nrow(vehicle), replace = TRUE, prob=c(0.7, 0.3))
#prob indicates split 
```

```r
# partition training and testing data from ind 
training <- vehicle[ind==1,]
testing <- vehicle[ind==2,]
```

```r
dim(testing)
```

```r
dim(training)
```

```r
cbind(summary(training$lc), summary(testing$lc))
```

```r
summary(training$lc)
```

```r
summary(testing$lc)
```

```r
#Create Multiple Linear Regresion with R 
model <- lm(lc~lh+Mileage, data=training)
```

```r
(model)
```

```r
# From the model we can see lh has a greater relationship and 
# it is positive. 
```

```r
summary(model)
```

```r
# Pr when less that 0.5 the variable is significant 
# .506 for Mileage should not be used. 
```

```r
model1 <- lm(lc ~ lh, data = training)
```

```r
summary(model1)
```

```r
# Our R-squared is 0.9446, we can say that variablility seen 
# in lh is explained by 94% in variablity in lh 
# p-value ; indicate the significance of the entire model 
```

```r
plot(lc~lh, training)
abline(model, col = "blue")
```

```r
# Model Diagnostics 
# Diagnostics & Prediction 
par(mfrow=c(2,2))     #plot 2 plots into rows 
plot(model)

```

### Residual vs Fitted 
 Used to detect non-linearity, unequal error variances and outliers 
 We look out for 3 things in this plot 
 1. The red line should be horiziontal to the x axis and be close to zero. 
 2. The dots should be scattered around the red lines 
 3. Shows us our outliers 
 
### Normal Q-Q
Ideal situation would be when all dots line up on the line 

### Scale Location
Check if residuals are spread equally along the range of predictors, Homoskedacity 

### Residuals vs Leverage 
To look for outlier in upper & lower right corner

### Multicollinearity
Occurs when there is moderate to high inter-correlations among independent variable. 
Leads to unstable estimates. 
Variance inflation factor(VIF) > 10 indicates presense of multicollinearlity 
When we have two variable which are highly correlated we do not need to add the two variables. 

We can check with the VIF. If VIF>10 then we have multicollinearity
