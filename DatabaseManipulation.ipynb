## Connect to MongoDB Atlas Database 




ocllection: name of the collection to connect to. Defaults to "test".
db: name of the database to connect to. Defaults to "test".
url: address of the MongoDB server in standard URI Format.
verbose: if TRUE, emits some extra output
options: additional connection options such as SSL keys/cert

```r

# Set the URL path with username, password and cluster url. 
# url_path = 'mongodb+srv://<username here>:<password here>!@<cluster url>/admin'


# using config to keep credientials safe, import the url 
url_path = config::get("url")

# Connect to the collection Ni-Based-Alloys from the Database Ni-Alloys using the above url path
# Verbose true would help display both warning and info 
 Ni <- mongo(collection = "Ni-Based-Alloys", db = "Ni-Alloys", 
               url = url_path, 
               verbose = TRUE)


print(Ni)

```

# Query Data

###  Find specific documents

Find Alloys with the composition of Tungsten W as 2.1 

```r

Ni$find('{"W": 2.1}')
```

Find alloys with the composition of Cobalt Co as 7.0

```r
Ni$find('{"Co": 7.0}')
```

### Find documents that contain Al, sorting from the lowest to highest
 <br>
 sort = '{"Al": 1}' indicates we sort from the lowest to highest 
 <br>
 sort = '{"Al": -1}' indicates we sort from the highest to lowest 
 


```r
# Ni$find('{"Al": {"$exists" : true}}')
# Ni$find('{"Al": {"$ne" : null}}')


 Ni$find(
  query = '{"Al": {"$exists" : true}}', 
  sort = '{"Al": 1}',
    limit = 5
     
)
```

### Find Documents that contain a specific element 
Find Alloys that contains Al, displaying the selected fields we are interested in

```r
# Ni$find('{"Al": {"$exists" : true}}')
# Ni$find('{"Al": {"$ne" : null}}')


 Ni$find(
  query = '{"Al": {"$exists" : true}}', 

      fields = '{"Alloy" : true, "Ni": true, "Co" : true,"Cr":true, "Mo": true, "W" : true, "Al": true, "Ta": true, "Source": true, "Table":true}',
  limit = 5
)
```

### Find Documents that exclude a particular property / element

Find Alloys that do not contain Hf

```r
 Ni$find(
  query = '{"Hf": {"$exists" : false}}', 

  limit = 5
)
```

### Find Documents according to Author / Paper 

Find documents from the paper 2010Feng

```r
Ni$find(
    query = '{"Source": "2010Feng"}',
    limit = 5
)

```

### Find documents with combination of queries using $in 
Find documents from the following papers 2011Fend, 2010Feng and 2015Li

```r
Ni$find(
    query = '{"Source" : { "$in" : [ "2011Feng", "2010Feng", "2015Li" ] } }',
    limit = 5
    )
```

## Find Documents from Keywords 

### Find documents relating to "phase"  from Tables key
Here we use regex which is a regular expression that defines a search pattern.
<br>
$options : i indicates the seach wouldn't be case-sensitive

```r
Ni$find(
    query = '{ "Table" : { "$regex" : "Phase", "$options" : "i" }  }',
    limit = 5
)
```

### Find documents relating to "comp" (compostion)  from Tables key


```r
Ni$find(
    query = '{ "Table" : { "$regex" : "comp", "$options" : "i" }  }',
    limit = 5
)
```

## Year Published Search Enter range of year published to search.

## Combining Queries 

### Required Elements, Display Alloys  which have the following Elements 
Ni, Co, Cr, Mo, W, Al, Ta



```r
Ni$find(
  query = '{"Ni" : { "$exists" : true }, "Co" : { "$exists" : true }, "Cr" : { "$exists" : true }, "Mo" : { "$exists" : true }, "W" : { "$exists" : true }, "Al" : { "$exists" : true }, "Ta" : { "$exists" : true }}', 
  fields = '{"Alloy" : true, "Ni" : true, "Co" : true,"Cr":true, "Mo": true, "W" : true, "Al": true, "Ta": true, "Source": true, "Table": true}',
  limit = 5
)
```

## Optional Elements, display Alloys which have either of the following elements 


Al and Hf 

```r
 Ni$find(
  query = '{"$or":[ {"Al" : { "$exists" : true }}, {"Hf" : { "$exists" : true }} ] }', 
  fields = '{"Alloy" : true, "Al" : true, "Hf": true}',
  limit = 5
)
```

## Excluded Compounds, find alloys that do not contain the following Elements 

Co, Cr and Mo

```r
Ni$find(
  query = '{"Co" : { "$exists" : false }, "Cr" : { "$exists" : false}, "Mo" : { "$exists" : false } }', 
  fields = '{"Alloy" : true, "Ni" : true, "Co" : true,"Cr":true, "Mo": true, "W" : true, "Al": true, "Ta": true }',
  limit = 5
)
```

## Query Alloy Composition 
Find Alloys that contain either Al of 0.53 wt% or Mo 2.0 wt%

```r
Ni$find('{"$or": [ {"Al" : 0.53}, {"Mo" : 2.0}  ]}')
```

## Update and  Upsert a record
The update method has the following arguments 
upsert=TRUE : Creates a new key and value pair if the set pairs are not available 
\b
multiple = TRUE: To update multiple documents, use the multi option in the update() method.

```r
Ni$update('{"Alloy":"TMS-2004"}', '{"$set":{"Hf":  20}}')
```

```r
Ni$update('{"Alloy":"TMS-2004"}', '{"$set":{"Author":  "Murakumo"}}', upsert=TRUE)
```

```r
Ni$find('{"Alloy": "TMS-2004"}')
```

## Remove the Document from the collection
just_one = TRUE argument indicates removal of just one record if multiple records are found

```r

Ni$remove('{"Alloy": "TMS-2004"}', just_one=TRUE)
Ni$count('[]')
```
